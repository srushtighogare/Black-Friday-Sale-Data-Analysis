## Outlier removal using IQR

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("Datasets/scholarship.csv")
```

```python
data.head()
```

```python
##plot graph 
sns.distplot(data["scholarship_exam_marks"])
```

```python
sns.boxplot(data["scholarship_exam_marks"])
```

```python
## to find the quartile using quantile function
Q1 = data["scholarship_exam_marks"].quantile(0.25)
Q3 = data["scholarship_exam_marks"].quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
## find the inter-quartile range
IQR = Q3 - Q1
```

```python
## IQR can also be found by the describe function
data.describe()
```

```python
## find minimum and maximum value 
min = Q1 - 1.5 * IQR
max = Q3 + 1.5 * IQR
```

```python
min
```

```python
max
```

```python
data[(data["scholarship_exam_marks"] < min) | (data["scholarship_exam_marks"] > max)]
```

## Trimming

```python
new_data = data[(data["scholarship_exam_marks"] >min) & (data["scholarship_exam_marks"] < max)]
```

```python
plt.figure(figsize = (16,8))
plt.subplot(2,2,1)
sns.distplot(data["scholarship_exam_marks"])

plt.subplot(2,2,2)
sns.boxplot(data["scholarship_exam_marks"])

plt.subplot(2,2,3)
sns.distplot(new_data["scholarship_exam_marks"])

plt.subplot(2,2,4)
sns.boxplot(new_data["scholarship_exam_marks"])

plt.show()
```

## Capping

```python
data["scholarship_exam_marks"] = np.where(data["scholarship_exam_marks"] < min, min , 
                                          np.where (data["scholarship_exam_marks"] > max, max , data["scholarship_exam_marks"] ))
```
