# Analyzing Gender

```python
#Approach 1
data = pd.DataFrame({"Ratio" : [len(df[df["Gender"] == "M"]) , len(df[df["Gender"] == "F"])]}, index = ["Male" , "Female"])
data.plot(y = "Ratio" , kind = "pie" , autopct = "%.2f" , figsize = (6,5))
```

```python
# Approach 2
df.groupby("Gender").size().plot(kind = "pie" , 
                                 autopct = "%.2f" , 
                                 figsize = (6,5),
                                title = "Gender Ratio")
```

```python
df.groupby("Gender").size().plot(kind = "bar" , 
                                 figsize = (6,5),
                                title = "Gender Ratio")
```

```python
# Ratio of purchases made by male and female

df.groupby("Gender").sum()["Purchase"].plot(kind = "pie" , autopct = "%.2f")
```

```python
#Average spend by each

df.groupby("Gender")["Purchase"].mean().plot(kind = "pie" , autopct = "%.2f")
```

# Analysing Age and Martial Status

```python
df.head()
```

```python
# For number of products they have purchased

df.groupby("Age").size().plot(kind = "bar" , title = "Purchase Distribution by Age" , figsize = (9,4))
```

```python
for i in df["Age"].unique():
    print(i, "\t:" , len(df[df["Age"] == i ]))
    
```

```python
df[df["Age"] == i ]["Product_ID"].nunique()
```

```python
#Uniques for each age

lst = []
for i in df["Age"].unique():
   lst.append((i , df[df["Age"] == i ]["Product_ID"].nunique()))
data = pd.DataFrame(lst , columns = ["Age" ,"Products"])

data.plot(kind = "bar" , x = "Age" )
```

```python
# Total amount spend according to age 

df.groupby("Age").sum()['Purchase'].plot(kind = "bar" , title = "Amount spent by Age" , figsize = (9,4))
```

```python
# Average amount spent 

df.groupby("Age")['Purchase'].mean().plot(kind = "pie" , title = "Amount spent by Age" , figsize = (9,4) , autopct = "%.1f")
```

```python
#Based on Marital Status

df.groupby("Marital_Status").size().plot(kind = "pie" , title = "Amount spent by Age" , figsize = (9,4) , autopct = "%.1f")
```

# Analysis Combining Gender and Marital Status

```python
df.head()
```

```python
# this code to add a new column where we ccome to know how many of the genders are married and not

l = []
for i in range(len(df)):
    l.append(df['Gender'][i] +"_"+ str(df['Marital_Status'][i]))

df['MaritalGender'] = l
```

```python
df.head()
```

```python

sns.countplot(x = df['MaritalGender'])
```

```python
sns.countplot(x = df['Age'], hue = df['MaritalGender'])
```

```python
sns.countplot(x = df['Product_Category_1'], hue = df['MaritalGender'])
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'], hue = df['MaritalGender'])
```

```python
sns.countplot(x = df['City_Category'], hue = df['MaritalGender'])
```
