```python
import pandas as pd
import seaborn as sns
```

## Categorical and Numerical

```python
df = sns.load_dataset("titanic")
df.head()
```

## Bar graph


```python
sns.barplot(x = "class" , y = "age" , data = df)
```

```python
sns.barplot(x = "class" , y = "fare" , data = df, estimator = "sum" , hue = "sex")
```

```python
sns.boxplot(x = "class" , y = "age" , data = df)
```

```python
sns.boxplot(x = "class" , y = "age" , data = df, hue = "survived")
```

```python
sns.violinplot(x = "class" , y = "age" , data = df)
```

```python
sns.distplot(df[df["class"] == "Second"]["age"])
```

## Categorical

```python
sns.countplot(x = "survived", data = df, hue = "class")
```

## Heatmaps

```python
 x = pd.crosstab (df["survived"],df["class"]) # to conctruct a pivot table
```

```python
x
```

```python
sns.heatmap(x, annot = True, fmt = ".1f" , cmap = "coolwarm")
```

```python
## one numerical data with 2 categorical
sns.barplot(x = "class" , y = "age" , hue = "sex" ,data = df) 
```

```python
sns.stripplot(x = "class" , y = "age" , hue = "sex" ,data = df) 
```
