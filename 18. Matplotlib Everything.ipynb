### 1.1 Generating Sin and Cos Graph using Matplotlib

```python
# made a list of values starting from 0 going up to 3.14 and make equal intervals for it
x = np.linspace(0, 4*np.pi, 100) # defining the data from 0 to pi value, (start , end , no of intervals)

# Defining Wave - sin and cosine these both are periodic functions. 
sin_y = np.sin(x) # creation of sine wave
cos_y = np.cos(x)

# Plotting the Wave
plt.plot(x, sin_y, label = 'Sine') # (value x, sine wave for the particular x , label = to give)
plt.plot(x, cos_y, label = 'Cosine')

plt.legend(loc = 'upper right') # legend used to display the label
plt.title('Sine Wave')

plt.show() 
```

1. Legend is used to add labels to the graph given in the plot function. The loc argument takes an integer or string that specifies the location of the legend on the plot. different values for loc, such as lower right, center, upper right, lower left, etc.

2 . title function adds a title to the plot

'-' will construct a line like in above plots

```python

plt.plot(x, sin_y, '*',label = "sine") 
plt.plot(x, cos_y, '--',label = "cos")

plt.legend(loc = 'upper right') # legend used to display the label
plt.title('Sine Wave')

plt.show() 
```

Note

In an IPython notebook the graph will be plotted even without using the plt.show() but in some conditions when we are working on big datasets and want to find some inference we use the plt.show() so it is generally a good practice to use it and also if we are writing a python script then also it's necessary to use the plt.show() command

### 4. Different Types of Plots

A line plot is a type of plot used to visualize the relationship between two continuous variables. It is a plot that displays data points as connected line segments. Line plots are commonly used to visualize trends and patterns in data, especially when the data is continuous and has a large number of observations.

In a line plot, the x-axis represents one variable and the y-axis represents another variable. The data points are connected by lines to create a plot that shows how the values of the two variables change with respect to each other. Line plots are often used to visualize time-series data, where the x-axis represents time and the y-axis represents a measurement.

In Matplotlib, the plot function is used to create a line plot. You can specify the x and y values of the data points, along with other optional arguments such as color, line style, and markers . You can also add labels, titles, and legends to your line plot to provide context and clarity.

```python
# for line plot here we have defined the x and y axis, x is the data , and y is the sin wave for x
x = np.linspace(0,4*np.pi, 100)
y = np.sin(x)

plt.style.use('Solarize_Light2')

plt.plot(x, y)

plt.title('Sine & Cosine Wave')

plt.show()
```

#### 4.1.2 Multiple Line Plots

```python
x = np.linspace(0,4*np.pi, 100) # x is again the same main data on which we want to plot

y_sin = np.sin(x) # this is y1 , sin wave for x
y_cos = np.cos(x) # this is y2 , cos wave for x

plt.figure(figsize = (15,5)) # size for the graph figure
plt.style.use('Solarize_Light2') # style

plt.plot(x, y_sin,  '*' ,color = 'red' , linewidth = 1, label = 'sin') # parameters defined 
plt.plot(x, y_cos, '--' , color = 'blue', linewidth = 1, label = 'cos')

plt.title('Sine & Cosine Wave')
plt.legend()
plt.show()
```

### 4.2 Scatter Plots

A scatter plot is a type of data visualization in which individual data points are represented as points in a two-dimensional plane. The x-coordinate of each data point represents one feature of the data, and the y-coordinate represents another feature. Scatter plots are commonly used to visualize the relationship between two variables and to identify patterns or outliers in the data.

In Matplotlib, scatter plots can be created using the scatter function. The scatter function accepts two arrays of values, representing the x-coordinates and y-coordinates of the data points. You can also specify additional arguments to customize the appearance of the scatter plot, such as the size and color of the markers, the transparency of the markers, and the marker shape.

```python
n = 500
x = np.random.randn(n)
y = np.random.randn(n)

plt.scatter(x,y, c = 'red', alpha = .5 ) # c = color , alpha = transparency , s = size

plt.title('Scatter Plot')

plt.show()
```

```python
n = 1000

x1 = np.random.randn(n)
y1 = np.random.randn(n)

x2 = np.random.randn(n) + 2 # this is used to to define the clusters of each group. if you forget what this does, change the numbers and observe.
y2 = np.random.randn(n) + 2

plt.scatter(x1,y1, c = 'red',  alpha = .7, s = 10, label = 'Group - 1')

plt.scatter(x2,y2, c = 'blue', alpha = .7, s = 10, label = 'Group - 2')

plt.title('Scatter Plot with Multiple Groups')
plt.legend()
plt.show()
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100) # to differentiate we are going to define 100 different colors for different values
sizes  = rng.rand(100) * 1000 # and same for size. multiply with 1000 cause size so small. 

plt.scatter(x,y, c = colors, s = sizes , alpha = .7, cmap = 'viridis', edgecolors = 'black' )
# cmap is the color map , edge colors is the color for the edges of the bubbles 
plt.title('Random Scatter Plot')
plt.xlabel('x values')
plt.ylabel('y values')

plt.colorbar() # to show the color scale

plt.show()
```

### 4.3 Histograms and Binnings

A histogram is a plot that displays the distribution of a set of continuous or discrete data. The data is divided into a set of intervals or bins, and the height of each bar in the plot represents the number of data points that fall within that bin. Binnings refers to the process of dividing the data into these bins. The choice of bin size can greatly affect the appearance of the histogram, so it's important to choose a bin size that accurately represents the underlying distribution of the data.

```python
data = np.random.randn(1000)
plt.style.use('Solarize_Light2')
plt.hist(data , edgecolor = "black" )
plt.show()
#here no of bins is 10
```

```python
data = np.random.randn(10000000)

plt.hist(data, bins = 2000, edgecolor = 'blue', alpha = .8, color = 'blue')

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
# as we increase the no of values and the no of bins, we will observe that we get a best fit graph
```

Individual Bins can be removed using histtype='stepfilled'

```python
data = np.random.randn(1000)

plt.hist(data, bins = 20, edgecolor = 'Black', alpha = .8, color = 'red', histtype = 'stepfilled')

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

#### 4.3.2 Multiple histograms

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(1000) + 2 
x3 = np.random.randn(1000) + 4

plt.hist(x1, bins = 20, alpha = .8 , edgecolor = "black")
plt.hist(x2, bins = 20, alpha = .8 , edgecolor = "black")
plt.hist(x3, bins = 20, alpha = .8 , edgecolor = "black")

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

### 4.4 2D Histograms

plt.hist2d is a function in matplotlib that generates a 2D histogram. A 2D histogram is a representation of the distribution of data over two variables. In contrast to a regular histogram which shows the distribution of a single variable, a 2D histogram shows the relationship between two variables. The function creates a heatmap that shows the density of data points in a 2D space.

we get a top view of histogram with the help of this 2D function, and can get to know how dense the graph is in the center, towards the right and left

```python
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hist2d(x,y, bins = 20 , cmap = 'viridis')

plt.title('2D Histogram')
plt.ylabel('x')
plt.xlabel('y')

plt.show()
```

```python
plt.hist2d(x,y, bins = 20 , cmap = 'Blues')
plt.title('2D Histogram')
plt.ylabel('x')
plt.xlabel('y')
plt.colorbar()

plt.show()
```

### 4.5 Pie Charts
A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]

fig = plt.figure(figsize = (5,5)) # decide a fig size

plt.pie(data, labels = cars)

plt.show()
```

#### 4.5.2 Basic operations

Explode - the part where we want the graph a little outward , it will be seperated from the pie chart

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0]  # all other values ar 0 except for jaguar , which will be seperated from the chart

fig = plt.figure(figsize = (5,5))

plt.pie(data, labels = cars, explode = explode)

plt.show()
```

Adding Shadows

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 

fig = plt.figure(figsize = (5,5))

plt.pie(data, labels = cars, explode = explode, shadow = True ) # set to true

plt.show()
```

Adding Percentage

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   50  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 

fig = plt.figure(figsize = (5,5))

plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%') # autopct set accordingly

plt.show()
```

Adding Wedget

```python
fig = plt.figure(figsize = (5,5))

plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%',
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 2, 'antialiased' : True}) #antialised is more of a clear version , when True

plt.show()
```

```python
fig = plt.figure(figsize = (5,5))

plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%',
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 2, 'antialiased' : False}) #antialised is more of a clear version , when false

plt.show()
```

### 4.6 Multiple Subplots

Multiple subplots refer to dividing a figure into multiple subplots or axes, each of which can contain a separate plot. This allows for multiple plots to be displayed in a single figure. In matplotlib, you can create multiple subplots using the subplot function.

Index Based

```python
x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2) # assign rows and columns required 

y = np.sin(x) #for the first block
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')


y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')


y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')


plt.tight_layout() # for better output and clean arrangement
plt.show()
```

With Loops - Example 1

```python
plt.subplots_adjust(hspace = .4, wspace = .4)

for i in range(1,7):
    
    plt.subplot(2,3,i)
    plt.text(0.5 , 0.5, str((2,3,i)), fontsize = 18, ha = 'center')

plt.show()
```

With Loops - Example 2

```python
x = np.linspace(0,10,50)
y = np.sin(x)


fig , (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title('Sin Wave')
ax2.set_title('Sin Wave - Square')
ax3.set_title('Sin Wave - Cube')

plt.tight_layout()

plt.show()
```

### 4.7 3D plots

Matplotlib provides a toolkit for 3D plotting named mplot3d, which provides functionality to create 3D plots

```python
plt.axes(projection = '3d') # display of projection
plt.show()
```

```python
ax = plt.axes(projection = '3d')

z = np.linspace(0,15,100) # has one more side z , goes from 0-15
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x,y,z, 'blue')
plt.show()
```

```python
#reverse
ax = plt.axes(projection = '3d')

z = np.linspace(0,15,100) # has one more side z , goes from 0-15
x = np.cos(z)
y = np.sin(z)

ax.plot3D(x,y,z, 'blue')
plt.show()
```
