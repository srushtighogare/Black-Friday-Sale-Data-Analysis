## Different Methods of dealing with Missing Data
### 1. DELETING THE COLUMN WITH MISSING DATA
### 2. Deleting the row with missing data
### 3. Filling the missing values - Imputation
      ### (1) Numerical data - use mean
      ### (2) Categorial Data - 
                  ### use mode
                  ### assign the NaN values their own category
### 4. Advanced Imputation


```python
import pandas as pd
```

```python
df = pd.read_csv("Datasets/AB_NYC_2019.csv")
```

## Delete missing values

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.drop("last_review", axis = 1, inplace = True) ## deleting column wi
```

```python
df1.isnull().sum()
```

```python
## to remove the row 
df2 = df.copy()
```

```python
df2.isna().sum()
```

```python
df2.dropna(inplace = True)
```

```python
df2.isna().sum()
```

```python
df2.shape
```

## Filling the missing values using Imputation

```python
df3 = df.copy()
```

```python
df3.head()
```

```python
mean_value = df3["reviews_per_month"].mean()
df3["reviews_per_month"].fillna(mean_value)
```

```python
df3.head()
```

## Filling Categorical Data

```python
df3.nunique()
```

```python
df3["last_review"].fillna(df3["last_review"].value_counts().index[0]) # we are filling all th null values with mod 
# first it counts values occuring, how many times occuring and then it takes the index 0 
```

```python
df3.isna().sum()
```

```python
## dealing with missing values by making a new category
```

```python
df4 = df.copy()
```

```python
df4["last_review"].fillna("Not_reviewed")
```

```python
df4.sample(10)
```
